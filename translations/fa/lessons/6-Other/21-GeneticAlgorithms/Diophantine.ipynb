## تکلیف: معادلات دیوفانتین

> این تکلیف بخشی از [برنامه درسی هوش مصنوعی برای مبتدیان](http://github.com/microsoft/ai-for-beginners) است و از [این پست](https://habr.com/post/128704/) الهام گرفته شده است.

هدف شما حل معادله‌ای به نام **معادله دیوفانتین** است - معادله‌ای با ریشه‌های صحیح و ضرایب صحیح. به عنوان مثال، معادله زیر را در نظر بگیرید:

$$a+2b+3c+4d=30$$

شما باید ریشه‌های صحیح $a$,$b$,$c$,$d\in\mathbb{N}$ را پیدا کنید که این معادله را برآورده کنند.

نکات:
1. می‌توانید ریشه‌ها را در بازه [0;30] در نظر بگیرید.
1. به عنوان یک ژن، می‌توانید از لیست مقادیر ریشه‌ها استفاده کنید.
