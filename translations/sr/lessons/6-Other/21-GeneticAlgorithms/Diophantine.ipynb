## Задатак: Диофантске једначине

> Овај задатак је део [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) и инспирисан је [овим чланком](https://habr.com/post/128704/).

Ваш циљ је да решите такозвану **Диофантску једначину** - једначину са целобројним коренима и целобројним коефицијентима. На пример, размотрите следећу једначину:

$$a+2b+3c+4d=30$$

Потребно је да пронађете целобројне корене $a$,$b$,$c$,$d\in\mathbb{N}$ који задовољавају ову једначину.

Савети:
1. Можете разматрати корене у интервалу [0;30]
1. Као ген, размотрите коришћење листе вредности корена
