## Задание: Диофантовы уравнения

> Это задание является частью [курса "Искусственный интеллект для начинающих"](http://github.com/microsoft/ai-for-beginners) и вдохновлено [этой статьей](https://habr.com/post/128704/).

Ваша цель — решить так называемое **Диофантово уравнение** — уравнение с целыми корнями и целыми коэффициентами. Например, рассмотрим следующее уравнение:

$$a+2b+3c+4d=30$$

Необходимо найти целые корни $a$, $b$, $c$, $d\in\mathbb{N}$, которые удовлетворяют этому уравнению.

Подсказки:
1. Можно рассматривать корни в интервале [0;30].
1. В качестве гена можно использовать список значений корней.
