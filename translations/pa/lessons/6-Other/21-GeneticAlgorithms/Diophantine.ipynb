## ਅਸਾਈਨਮੈਂਟ: ਡਾਇਓਫੈਂਟਾਈਨ ਸਮੀਕਰਨ

> ਇਹ ਅਸਾਈਨਮੈਂਟ [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) ਦਾ ਹਿੱਸਾ ਹੈ ਅਤੇ ਇਹ [ਇਸ ਪੋਸਟ](https://habr.com/post/128704/) ਤੋਂ ਪ੍ਰੇਰਿਤ ਹੈ।

ਤੁਹਾਡਾ ਲਕਸ਼ ਹੈ ਕਿ ਤੁਸੀਂ ਇਸੇ-ਕਹੇ ਜਾਣ ਵਾਲੇ **ਡਾਇਓਫੈਂਟਾਈਨ ਸਮੀਕਰਨ** ਨੂੰ ਹੱਲ ਕਰੋ - ਇੱਕ ਸਮੀਕਰਨ ਜਿਸਦੇ ਜੜ੍ਹਾਂ ਅਤੇ ਗੁਣਾਂ ਪੂਰਨ ਅੰਕਾਂ ਵਿੱਚ ਹੁੰਦੇ ਹਨ। ਉਦਾਹਰਣ ਵਜੋਂ, ਹੇਠਾਂ ਦਿੱਤੇ ਸਮੀਕਰਨ ਨੂੰ ਵੇਖੋ:

$$a+2b+3c+4d=30$$

ਤੁਹਾਨੂੰ ਪੂਰਨ ਅੰਕਾਂ ਦੀਆਂ ਜੜ੍ਹਾਂ $a$,$b$,$c$,$d\in\mathbb{N}$ ਲੱਭਣੀਆਂ ਹਨ ਜੋ ਇਸ ਸਮੀਕਰਨ ਨੂੰ ਪੂਰਾ ਕਰਦੀਆਂ ਹਨ।

ਸੁਝਾਅ:
1. ਤੁਸੀਂ ਜੜ੍ਹਾਂ ਨੂੰ [0;30] ਦੇ ਅੰਤਰਾਲ ਵਿੱਚ ਮੰਨ ਸਕਦੇ ਹੋ।
1. ਇੱਕ ਜੀਨ ਵਜੋਂ, ਜੜ੍ਹਾਂ ਦੇ ਮੁੱਲਾਂ ਦੀ ਸੂਚੀ ਵਰਤਣ ਬਾਰੇ ਸੋਚੋ।



---

**ਅਸਵੀਕਰਤੀ**:  
ਇਹ ਦਸਤਾਵੇਜ਼ AI ਅਨੁਵਾਦ ਸੇਵਾ [Co-op Translator](https://github.com/Azure/co-op-translator) ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਅਨੁਵਾਦ ਕੀਤਾ ਗਿਆ ਹੈ। ਜਦੋਂ ਕਿ ਅਸੀਂ ਸਹੀ ਹੋਣ ਦੀ ਕੋਸ਼ਿਸ਼ ਕਰਦੇ ਹਾਂ, ਕਿਰਪਾ ਕਰਕੇ ਧਿਆਨ ਦਿਓ ਕਿ ਸਵੈਚਾਲਿਤ ਅਨੁਵਾਦਾਂ ਵਿੱਚ ਗਲਤੀਆਂ ਜਾਂ ਅਸੁਚੀਤਤਾਵਾਂ ਹੋ ਸਕਦੀਆਂ ਹਨ। ਮੂਲ ਦਸਤਾਵੇਜ਼ ਨੂੰ ਇਸਦੀ ਮੂਲ ਭਾਸ਼ਾ ਵਿੱਚ ਅਧਿਕਾਰਤ ਸਰੋਤ ਮੰਨਿਆ ਜਾਣਾ ਚਾਹੀਦਾ ਹੈ। ਮਹੱਤਵਪੂਰਨ ਜਾਣਕਾਰੀ ਲਈ, ਪੇਸ਼ੇਵਰ ਮਨੁੱਖੀ ਅਨੁਵਾਦ ਦੀ ਸਿਫਾਰਸ਼ ਕੀਤੀ ਜਾਂਦੀ ਹੈ। ਇਸ ਅਨੁਵਾਦ ਦੀ ਵਰਤੋਂ ਤੋਂ ਪੈਦਾ ਹੋਣ ਵਾਲੇ ਕਿਸੇ ਵੀ ਗਲਤਫਹਿਮੀ ਜਾਂ ਗਲਤ ਵਿਆਖਿਆ ਲਈ ਅਸੀਂ ਜ਼ਿੰਮੇਵਾਰ ਨਹੀਂ ਹਾਂ।
