## অ্যাসাইনমেন্ট: ডায়োফ্যান্টাইন সমীকরণ

> এই অ্যাসাইনমেন্টটি [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners)-এর অংশ এবং এটি [এই পোস্ট](https://habr.com/post/128704/) থেকে অনুপ্রাণিত।

আপনার লক্ষ্য হলো তথাকথিত **ডায়োফ্যান্টাইন সমীকরণ** সমাধান করা - একটি সমীকরণ যার মূল এবং সহগ উভয়ই পূর্ণসংখ্যা। উদাহরণস্বরূপ, নিচের সমীকরণটি বিবেচনা করুন:

$$a+2b+3c+4d=30$$

আপনাকে এমন পূর্ণসংখ্যার মূল $a$,$b$,$c$,$d\in\mathbb{N}$ খুঁজে বের করতে হবে যা এই সমীকরণটি পূরণ করে।

ইঙ্গিত:
1. আপনি মূলগুলোকে [0;30] পরিসরে বিবেচনা করতে পারেন
1. একটি জিন হিসেবে, মূল মানগুলোর তালিকা ব্যবহার করার কথা ভাবতে পারেন



---

**অস্বীকৃতি**:  
এই নথিটি AI অনুবাদ পরিষেবা [Co-op Translator](https://github.com/Azure/co-op-translator) ব্যবহার করে অনুবাদ করা হয়েছে। আমরা যথাসম্ভব সঠিকতার জন্য চেষ্টা করি, তবে অনুগ্রহ করে মনে রাখবেন যে স্বয়ংক্রিয় অনুবাদে ত্রুটি বা অসঙ্গতি থাকতে পারে। মূল ভাষায় থাকা নথিটিকে প্রামাণিক উৎস হিসেবে বিবেচনা করা উচিত। গুরুত্বপূর্ণ তথ্যের জন্য, পেশাদার মানব অনুবাদ সুপারিশ করা হয়। এই অনুবাদ ব্যবহারের ফলে কোনো ভুল বোঝাবুঝি বা ভুল ব্যাখ্যা হলে আমরা দায়বদ্ধ থাকব না।
