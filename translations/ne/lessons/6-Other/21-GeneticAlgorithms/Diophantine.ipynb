## असाइनमेन्ट: डायोफेन्टाइन समीकरणहरू

> यो असाइनमेन्ट [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) को हिस्सा हो र [यो पोस्ट](https://habr.com/post/128704/) बाट प्रेरित छ।

तपाईंको लक्ष्य तथाकथित **डायोफेन्टाइन समीकरण** समाधान गर्नु हो - जसमा पूर्णांक मूलहरू र पूर्णांक गुणांकहरू हुन्छन्। उदाहरणका लागि, तलको समीकरणलाई विचार गर्नुहोस्:

$$a+2b+3c+4d=30$$

तपाईंले यस्तो पूर्णांक मूलहरू $a$,$b$,$c$,$d\in\mathbb{N}$ पत्ता लगाउनुपर्छ जसले यो समीकरणलाई सन्तुष्ट पार्छ।

सुझावहरू:
1. मूलहरूलाई [0;30] को अन्तरालमा विचार गर्न सक्नुहुन्छ।
1. जीनको रूपमा, मूल मानहरूको सूची प्रयोग गर्न विचार गर्नुहोस्।



---

**अस्वीकरण**:  
यो दस्तावेज़ AI अनुवाद सेवा [Co-op Translator](https://github.com/Azure/co-op-translator) प्रयोग गरी अनुवाद गरिएको हो। हामी यथासम्भव सटीकता सुनिश्चित गर्न प्रयास गर्छौं, तर कृपया ध्यान दिनुहोस् कि स्वचालित अनुवादहरूमा त्रुटि वा अशुद्धता हुन सक्छ। यसको मूल भाषामा रहेको मूल दस्तावेज़लाई आधिकारिक स्रोत मानिनुपर्छ। महत्त्वपूर्ण जानकारीका लागि, व्यावसायिक मानव अनुवाद सिफारिस गरिन्छ। यस अनुवादको प्रयोगबाट उत्पन्न हुने कुनै पनि गलतफहमी वा गलत व्याख्याको लागि हामी जिम्मेवार हुने छैनौं।
