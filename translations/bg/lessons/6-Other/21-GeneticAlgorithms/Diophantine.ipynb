## Задача: Диофантови уравнения

> Тази задача е част от [Учебната програма за начинаещи в AI](http://github.com/microsoft/ai-for-beginners) и е вдъхновена от [тази публикация](https://habr.com/post/128704/).

Вашата цел е да решите така нареченото **Диофантово уравнение** - уравнение с цели корени и цели коефициенти. Например, разгледайте следното уравнение:

$$a+2b+3c+4d=30$$

Трябва да намерите цели корени $a$,$b$,$c$,$d\in\mathbb{N}$, които удовлетворяват това уравнение.

Подсказки:
1. Можете да разгледате корените в интервала [0;30]
1. Като ген, обмислете използването на списък с стойности на корените
