## Завдання: Діофантові рівняння

> Це завдання є частиною [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) і натхнене [цією статтею](https://habr.com/post/128704/).

Вашою метою є розв'язання так званого **Діофантового рівняння** – рівняння з цілими коренями та цілими коефіцієнтами. Наприклад, розглянемо наступне рівняння:

$$a+2b+3c+4d=30$$

Вам потрібно знайти цілі корені $a$,$b$,$c$,$d\in\mathbb{N}$, які задовольняють це рівняння.

Підказки:
1. Ви можете розглядати корені в інтервалі [0;30]
1. Як ген, розгляньте використання списку значень коренів
