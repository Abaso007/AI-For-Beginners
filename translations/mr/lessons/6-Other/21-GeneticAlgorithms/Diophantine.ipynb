## असाइनमेंट: डायोफंटाइन समीकरणे

> हे असाइनमेंट [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) चा भाग आहे आणि [या पोस्ट](https://habr.com/post/128704/) वरून प्रेरित आहे.

तुमचे उद्दिष्ट म्हणजे **डायोफंटाइन समीकरण** सोडवणे - एक असे समीकरण ज्यामध्ये पूर्णांक मुळे आणि पूर्णांक गुणांक असतात. उदाहरणार्थ, खालील समीकरण विचार करा:

$$a+2b+3c+4d=30$$

तुम्हाला असे पूर्णांक मुळे $a$,$b$,$c$,$d\in\mathbb{N}$ शोधायचे आहेत जे हे समीकरण पूर्ण करतात.

सूचना:
1. मुळे [0;30] या श्रेणीत असू शकतात असे गृहीत धरा.
1. जीन म्हणून, मूळ मूल्यांच्या यादीचा विचार करा.



---

**अस्वीकरण**:  
हा दस्तऐवज AI भाषांतर सेवा [Co-op Translator](https://github.com/Azure/co-op-translator) वापरून भाषांतरित करण्यात आला आहे. आम्ही अचूकतेसाठी प्रयत्नशील असलो तरी कृपया लक्षात ठेवा की स्वयंचलित भाषांतरे त्रुटी किंवा अचूकतेच्या अभावाने युक्त असू शकतात. मूळ भाषेतील दस्तऐवज हा अधिकृत स्रोत मानला जावा. महत्त्वाच्या माहितीसाठी व्यावसायिक मानवी भाषांतराची शिफारस केली जाते. या भाषांतराचा वापर करून उद्भवलेल्या कोणत्याही गैरसमज किंवा चुकीच्या अर्थासाठी आम्ही जबाबदार राहणार नाही.
